#### Import all the libraries that are necessary.

```python
import pandas as pd   # importing all necessary libraries
import numpy as np
```

#### Open the ` Employee.csv` file and name your dataset `employee`.

```python
employee = pd.read_csv("Employee.csv")   # opening the Employee.csv file and naming it employee
```

#### What's the mean salary by department?
Hint : There are 2 ways to do it - one way is with groupby and the other way is with pivot tables. Do it both ways. 


```python
#employee.groupby(['Department']).agg({'Salary':np.mean}).reset_index().head() # mean salary by department using groupby

employee.pivot_table(index= ['Department'], values = ['Salary'], aggfunc = {'Salary':['mean'],}).reset_index()
```

#### We want now to group the data by more than one column. Use Pandas pivot table function to select the mean salary by department and title.

```python
employee.groupby(['Department', 'Title']).agg({'Salary':np.mean}).reset_index().head() # mean salary by department and title using pivote table
#employee.pivot_table(index= ['Department', 'Title'], values = ['Salary'], aggfunc = {'Salary':['mean'],}).reset_index() 
```

#### We want to display the number of years of work experience and the salary for each Title. Add Years as a column in your pivot table.
*Hint: Use Pandas doc(see References in README.md) to explore the pivot_table() arguments*

```python
emp1 = employee.pivot_table(index= ['Title'], values = ['Salary','Years'], aggfunc = {'Salary':['sum'],'Years':'sum'}).reset_index() # number of years of work experience and the salary for each Title
emp1.head()
```

#### We want to know the information about the mean salary and number of employees in each department and for each job title.

```python
# information about the mean salary and number of employees in each department and for each job title
emp2 = employee.pivot_table(index= ['Department', 'Title'], values = ['Salary','Name'], aggfunc = {'Salary':['mean'],'Name':'count'}).reset_index() # number of years of work experience and the salary for each Title
emp2.head()
```

#### We want to know for each department and for each job title the Salary's median, minimum, maximum and standard deviation. Comment your results.

```python
emp3 = employee.pivot_table(index= ['Department', 'Title'], values = ['Salary'], aggfunc = {'Salary':['median', 'min', 'max', 'std']}).reset_index() # median, minimum, maximum and standard deviation for each department and for each job title for the Salary
emp3.head() 
```

#### Based on your comments, fill the missing values with an appropriate value.

```python
emp4 = employee.pivot_table(index= ['Department', 'Title'], values = ['Salary'], aggfunc = {'Salary':['median', 'min', 'max', 'std']}, fill_value=0).reset_index() # output contains missing values, the solution to deal with missing values is shown below 
emp4.head() 
```

#### The stake holders want to know for each department the number of employees and how much money is spend on salaries. Could you provide that information? 

```python
emp5 = employee.pivot_table(index= ['Department'], values = ['Salary','Name'], aggfunc = {'Salary':['sum'],'Name':'count'}).reset_index() # the number of employees and how much money is spend on salaries for each department
emp5.head()
```

#### For each Department and Title, we want to know the the total years of work experience and the mean salary.

```python
emp6 = employee.pivot_table(index= ['Department'], values = ['Salary','Years'], aggfunc = {'Salary':['mean'],'Years':'count'}).reset_index() # total years of work experience and the mean salary for each Department
emp6.head()
```

#### We now want to compute the mean salary after removing the maxima for each Department. Create the appropriate pivot table.

Hint: Write a custom function.

```python
dropMax = employee.apply(lambda col:col[col < col.max()])
dropMax
```

```python
emp7 = dropMax.pivot_table(index= ['Department'], values = ['Salary'], aggfunc = {'Salary':['mean']}).reset_index() # total years of work experience and the mean salary for each Department
emp7.head()
```
